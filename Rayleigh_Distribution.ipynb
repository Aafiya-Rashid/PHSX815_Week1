```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.rayleigh(1, 10000)
np.savetxt("rayleigh_random.txt", data)

f = np.loadtxt("rayleigh_random.txt")
plt.hist(f, bins = 100, color = "m", alpha = 0.75, edgecolor = 'k')
plt.title("Histogram of Rayleigh Distribution")
plt.grid(color = 'k', alpha = 0.3, ls = 'dotted', lw = 0.7)
plt.xlabel("x")
plt.ylabel("Counts")
plt.savefig('Rayleigh_Distribution.pdf')
plt.show()
```
